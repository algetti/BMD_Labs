```python
import pandas as pd
import numpy as np
import scipy.stats as sts
```

<strong><h3>1.Сгенеруйте набір даних, що містить не менше 6 змінних. Частину змінних згенеруйте за нормальний розподілом, частину - за будь-яким іншим розподілом. Кількість спостережень - 20.</strong></h3>

```python
data_fr = pd.DataFrame({
    'Player1': np.random.randn(20),
    'Player2':np.random.standard_normal(size = 20),
    'Player3':np.random.normal(loc = 10, scale = 3, size = 20),
    'Skills':np.random.poisson(lam = 5, size = 20),
    'Uniform':np.random.uniform(size = 20),
    'Variety':np.random.exponential(scale = 1, size = 20)    
})

data_fr
```

<strong><h3>2.Створіть функцію, що реалізує критерій Шапіро-Уілка та застосуйте її до кожної змінної набору данних. Детальний опис роботи критерію та приклад його розрахунку.</strong></h3>

```python
data_fr = pd.DataFrame({key: sorted(values) for key, values in data_fr.transpose().iterrows()})
```

```python
def data_mean(arr,n):
    all_data = 0
    for i in range(n):
        all_data += arr[i]
    return all_data/n
data_mean_value = {i: data_mean(data_fr[i], 20) for i in data_fr}
```

```python
def data_squared(arr,n, dat_mean):
    s = 0
    for i in range(n):
        s += ((arr[i]-dat_mean)**2)
    return s
data_squared_value = {i: data_squared(data_fr[i], 20, data_mean_value[i]) for i in data_fr}
```

```python
a = [0.4734, 0.3211, 0.2565, 0.2085, 0.1686, 0.1334, 0.1013, 0.0711, 0.0422, 0.0140]

def b(arr, n):
    B = 0
    for i in range(n//2):
        B += (a[i]*(arr[(n-1)-i] - arr[i]))
    return B

b_value = {i: b(data_fr[i], 20) for i in data_fr}
```

```python
def W (squared, b):
    return (b**2)/squared

W_value = {i: W(b_value[i],data_squared_value[i]) for i in data_fr}

WTheory = sts.shapiro(data_fr['Player1'])
W_value, WTheory
```

<strong><h3>3.Сгенеруйте пусту таблицю, в якій кількість рядків - це кількість змінних + 1. Кількість стовбчиків - 3. Назва стовпчиків - Variable, Statistics, Distribution.</strong></h3>

```python
column_name = ["Variable", "Statistics", "Distribution"]
table = pd.DataFrame(columns = column_name)
table
```

<strong><h3>4.Занести в стовпчик Variable назви змінних.</strong></h3>

```python
table["Variable"] = data_fr.keys()
table
```

<strong><h3>5. Занести в стовпчик Distribution результати роботи криетрію Шапіро-Уілка. Для змінних, які розподілені за нормальним законом вносити "normal", в протилежному вигляді - "non-normal".</strong></h3>

```python
table["Distribution"] = [('normal' if 'Norm' in i else 'non-normal') for i in table["Variable"]]
table
```

<strong><h3>6.Розрахувати для всіх змінних, що розподілені за нормальним законом розрахувати середне значення та стандартне відхилення. Записати в таблицю. Форма запису: Середнє [стандартне відхилення].</strong></h3>

```python
for i in range(table.shape[0]):
    if 'Norm' in table["Variable"][i]:
        table['Statistics'][i] = f'{data_fr[table["Variable"][i]].mean()} {[data_fr[table["Variable"][i]].std()]}'
table
```

<strong><h3>7.Розрахувати для всіх змінних, що розподілені за розподілом відсутнім від нормального розрахувати медіану та інтерквартильний розмах. Записати в таблицю. Форма запису: Медіана [25%; 75%].</strong></h3>

```python
for i in range(table.shape[0]):
    if 'Norm' not in table["Variable"][i]:
        table["Statistics"][i] = f'{data_fr[table["Variable"][i]].median()} [{data_fr[table["Variable"][i]].quantile(0.25)}; {data_fr[table["Variable"][i]].quantile(0.75)}]'
table
```

<strong><h3>8.Зберегти таблицю в пдф форматі</div></strong></h3>

```python
import matplotlib.backends.backend_pdf
import matplotlib.pyplot as plt
fig = plt.figure()
ax=fig.add_subplot(111)

cell_text = []
for row in range(len(table)):
    cell_text.append(table.iloc[row])

ax.table(cellText=cell_text, colLabels=table.columns, loc='center')
ax.axis('off')

pdf = matplotlib.backends.backend_pdf.PdfPages("table.pdf")
pdf.savefig(fig)
pdf.close()
```
